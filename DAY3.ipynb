## Remove duplicate elements from list and take list by user

```python
newlist=[]
l=[]
size=int(input("Enter a size: "))
for i in range(0,size):
    newlist.append(int(input()))

print("Original list:",newlist)
for i in newlist:
    if i not in l:
        l.append(i)
print("Updated list: ",l)
```

```python
newlist=[]
size=int(input("Enter a size: "))
for i in range(0,size):
    newlist.append(int(input()))
print("Original list:",newlist)

for i in newlist:
    if newlist.count(i)>1:
        newlist.remove(i)
print("Updated list:",newlist)
```

```python
newlist=[]
even=[]
odd=[]
size=int(input("Enter a size: "))
for i in range(0,size):
    newlist.append(int(input()))

print(newlist)
even.append(newlist[0])
for i in range(1,size):
    if i%2==0:
         even.append(newlist[i])
    else:
        odd.append(newlist[i])

even.sort()
odd.sort()
print(f"Even list={even}")
print(f"Odd list={odd}")
print(f"Second largest element of even index = {even[len(even)-2]}")
print(f"Second smallest element of odd index = {odd[1]}")
```
