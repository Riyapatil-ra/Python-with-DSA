## Remove duplicate elements from list and take list by user

```python
newlist=[]
l=[]
size=int(input("Enter a size: "))
for i in range(0,size):
    newlist.append(int(input()))

print("Original list:",newlist)
for i in newlist:
    if i not in l:
        l.append(i)
print("Updated list: ",l)
```

```python
newlist=[]
size=int(input("Enter a size: "))
for i in range(0,size):
    newlist.append(int(input()))
print("Original list:",newlist)

for i in newlist:
    if newlist.count(i)>1:
        newlist.remove(i)
print("Updated list:",newlist)
```
