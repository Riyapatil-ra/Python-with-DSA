## Inheritance

```python
class demo:
    def add(self,a,b):
        return a+b
class claculations(demo):
    def sub(self,a,b):
        return abs(a-b)

obj=demo()
cal=claculations()
print(cal.add(10,20)) # although cal object of calculations function don't have add function but it can inheritant it parent class i.e. demo class using inhreitance
print(cal.sub(30,40))
print(obj.add(60,70))
# print(obj.sub(60,10)) # since object of demo class don't have sub function and it is parent class.so it will show error.
```

### Multiple inheritance

```python
class demo:
    def add(self,a,b):
        return a + b
class demo1:
    def mul(self,a,b):
        return a * b
class claculations(demo,demo1):
    def sub(self,a,b):
        return abs(a-b)
    
obj=demo()
obj1=demo1()
cal=claculations()
print(cal.sub(90,30))
print(cal.mul(2,4))
print(cal.add(30,60))
# print(obj1.add(3,2))
```

### Multi-line inheritance

```python
class demo:
    def add(self,a,b):
        return a+b
class demo1(demo):
    def mul(self,a,b):
        return a*b
class claculations(demo1):
    def sub(self,a,b):
        return abs(a-b)
    
obj=demo()
obj1=demo1()
cal=claculations()
print(cal.sub(90,30))
print(cal.mul(2,4))
print(cal.add(30,60))
print(obj1.add(3,2))
```

## Polymorphism

```python
class sipna:
    def add(self,a,b):
        return a+b
class sipna1(sipna):
    def add(self,a,b,c):
        return a+b+c
obj=sipna()
obj1=sipna1()
print(obj.add(10,20))
print(obj1.add(10,20,30))
```

```python
class sipna:
    def add(self):
        print("sipna")
class sipna1(sipna):
    def add(self):
        print("sipna1")
        
class sipna2(sipna1):
    def add(self):
        print("sipna2")

obj=sipna()
obj1=sipna1()
obj2=sipna2()

for i in obj,obj1,obj2:
    print(i)
```

```python
class integer:
    def add(self,a,b):
        return a + b
class str1:
    def add(self,a,b):
        return a+b
class lit:
    def add(self,a,b):
        result=[]
        if len(a)!=len(b):
            print("0")
        else:
            for i in range(len(a)):
                result.append(a[i]+b[i])
            print(result)

obj1=integer()
obj2=str1()
obj3=lit()

print(obj1.add(10,20))
print(obj2.add("M","Y"))
print(obj3.add([10,20,30],[40,50,60]))
```

## Sorting

### Bubble Sort

```python
n=int(input("Enter a size:"))
l=list(map(int,input().split()))
print("Original list:",l)
for i in range(n):
    # print(f"Pass {i}:")
    for j in range(0,n-1):
        # print(f"IT {j}:",end="")
        # print(l)
        if (l[j+1]<l[j]):
            temp=l[j]
            l[j]=l[j+1]
            l[j+1]=temp

print("Updated list:",l)
            
```

### Optimize code for Bubble sort

```python
n=int(input("Enter a size:"))
l=list(map(int,input().split()))
print("Original list:",l)
for i in range(n):
    flag=False
    print(f"Pass {i}:")
    for j in range(0,n-1):
        print(f"IT {j}:",end="")
        print(l)
        if (l[j+1]<l[j]):
            flag=True
            temp=l[j]
            l[j]=l[j+1]
            l[j+1]=temp
    if flag==False:
        break
print("Updated list:",l)
```

### Selection Sort . Selection sort is "in-place" 

```python
n=int(input("Enter a size:"))
l=list(map(int,input().split()))
print("Original list:",l)
for i in range(0,n-1):
    min=i
    for j in range(i+1,n):
        if (l[min] > l[j]):
            min=j
    temp=l[i]
    l[i]=l[min]
    l[min]=temp

print("Updated list:",l)
```
