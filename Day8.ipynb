# Queue

## Linear Queue

```python
class Queue():
    def __init__(self):
        self.front=-1
        self.rear=-1
        self.size=5
        self.queue=[0]*self.size
    
    def Enque(self,data):
        if self.rear==self.size-1:
            print()
            print("Overflow")
            print()
        elif self.front==-1 and self.rear==-1:
            self.front+=1
            self.rear+=1
            self.queue[self.rear]=data
        else:
            self.rear+=1
            self.queue[self.rear]=data

    def Deque(self):
        if self.front==self.size:
            print()
            print("Underflow")
            print()
        else:
            print()
            print(f"{self.queue[self.front]} item deque")
            print()
            self.front+=1

    def Display(self):
        print()
        print("Queue elements are:",end=" ")
        for i in range(self.front,self.rear+1):
            print(self.queue[i],end=" ")
        print()

q=Queue()
choice=1
while(choice!=0):
    print("1.Adding elements")
    print("2.Deleteing element")
    print("3.Display")
    choice=int(input("Enter choice:"))
    if choice == 1:
        data=int(input("Enter data:"))
        q.Enque(data)
    elif choice == 2:
        q.Deque()
    elif choice == 3:
        q.Display()
```

# Searching

## Linear Searching

```python
def linear(arr,target):
    for i in range(0,n):
        if arr[i]==target:
            return i
    return -1

n=int(input("Enter size:"))
arr=list(map(int,input().split()))
target=int(input("Enter a target:"))
result=linear(arr,target)
if result==-1:
    print(f"Target {target} is not found in {arr}")
else:
    print(f"Target {target} is found at {result+1} in {arr}")
```

## Binary Searching

```python
def Binary(arr,target,low,high):
    if low<=high:
        mid=(low+high)//2
        if arr[mid]<target: # 24<50
            return Binary(arr,target,mid+1,high)
        elif arr[mid]>target: #67>50
            return Binary(arr,target,low,mid-1)
        elif arr[mid]==target:
            return mid
    return -1
  
n=int(input("Enter size:"))
arr=list(map(int,input().split()))
target=int(input("Enter a target:"))
arr.sort()
result=Binary(arr,target,0,n-1)
if result==-1:
    print(f"Target {target} is not found in {arr}")
else:
    print(f"Target {target} is found at {result+1} in {arr}")
```

# Mini Game

```python
import random as rd

num=rd.randint(1,100)
chances=9
while chances>0:
    choice=int(input("Guess the number:"))
    if choice < num:
        print(f"You guess too low !! you have only {chances} available")
        chances-=1
    elif choice > num:
        print(f"You guess too high !! you have only {chances} available")
    else:
        print("Congratulations you guess the number")
        break
    chances-=1
```

# Tree

```python
class node:
    def __init__(self,data):
        self.left=None
        self.data=data
        self.right=None

class Tree:
    def __init__(self):
        self.root=None

    def createNode(self,data):
        newnode=node(data)
        self.insert(self.root,newnode)

    def insert(self,root,data):
        if self.root==None:
            newnode=self.root
            return
        
        if root.left==None:
            root.left==newnode
        else:
            return self.insert(root.left,newnode)
        
        if root.right==None:
            root.right==newnode
        else:
            return self.insert(root.right,newnode)
        
t=Tree()
t.createNode(10)
t.createNode(20)
t.createNode(30)
t.createNode(40)
t.createNode(50)
```
