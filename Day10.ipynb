```python
s="ab#kc##jhkl##g"
print("Original string:",s)
temp1=""
temp2=""
for i in s:
    if i=='#':
        temp1+=i
    else:
        temp2+=i
result=temp1 + temp2
print("Updated string:",result)
```

```python
arr=[1,2,3]
def subset(index,result):
    if index == len(arr):
        print(result)
        return
    subset(index+1, result + [arr[index]])

    subset(index+1, result)

subset(0,[])
```

```python
coins=[10,5,2,1]
amount=18
result=[]
for i in coins:
    while amount>=i:
        result.append(i)
        amount-=i

print(result)
```

```python
data={}
def register():
    rollno=int(input("Enter roll no:"))
    name=input("Enter your name:")
    age=int(input("Enter your age:"))
    marks=0
    if rollno not in data:
        data[rollno]=[name,age,marks]

def Display():
    rollno=int(input("Enter roll no:"))
    if rollno in data:
        print(f"Name -> {data[rollno][0]}")
        print(f"Age -> {data[rollno][1]}")
        print(f"Marks -> {data[rollno][2]}")

def Search():
    rollno=int(input("Enter roll no:"))
    if rollno not in data:
        print("Student is not in database")
    else:
        print("Student is in database")

def marks():
    rollno=int(input("Enter roll no:"))
    marks=data[rollno][2]
    print(f"Marks of this student is {marks}")
    
def update():
    rollno=int(input("Enter rollno to be updated:"))
    marks=int(input("Enter marks to be updated:"))
    if not data[rollno]:
        print("Data no exist")   
    data[rollno][2]=marks
    print("Data is updated")

choice=1
while choice!=0:
    print("--------------Student Management System--------------")
    print("1.Register")
    print("2.Display")
    print("3.Search")
    print("4.marks")
    print("5.Update")
    choice=int(input("Enter your chocie:"))
    if choice==1:
        register()
    elif choice==2:
        Display()
    elif choice==3:
        Search()
    elif choice==4:
        marks()
    elif choice==5:
        update()
        
```
