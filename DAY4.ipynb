# Functions

```python
def fun(start,end,num):
    sum1=0
    sum2=0
    for i in range(start,end+1):
        if i % num==0:
            sum1=sum1+i
        else:
            sum2=sum2+i
    return sum1-sum2

start=int(input())
end=int(input())
num=int(input())

result=fun(start,end,num)
print("Difference : ",abs(result))
```

```python
def funy(s1):
    l=" "
    for i in s1:
        if i not in l:
            if s1.count(i)>1:
                l+=i
                l+=str(s1.count(i))
            else:
                l+=i
    print("Result:",l)

s1=input()
print("String:",s1)
funy(s1)
```

```python
def demo(s):
    s1=""
    for i in s:
        if i not in s1:
            if s.count(i)>1:
                s1+=i
                s1+=str(s.count(i))
            else:
                s1+=i
    print(s1)
s=input()
print(s)
demo(s)
```

## Positional arguments

```python
def demo(*args):
    c=0
    for i in args:
        c+=i
    return c
print(demo(10,20,30))
```

## Keyword argument

```python
def demo(**kwargs):
    return kwargs

print(demo(a='Riya',b='Patil'))
```

```python
n=int(input())
rev=0
sum=0
while n!=0:
    rev=n % 10
    sum+=rev
    n//=10
if sum>10:
    c=sum % 10
    sum=c+sum//10

print(sum)
```
