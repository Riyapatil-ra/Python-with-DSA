### Insert at end

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None

class linkedlist:
    def __init__(self):
        self.head = None

    def InsertAtbeg(self, data):
        newnode = node(data)
        if self.head == None:
            self.head = newnode
            newnode.next=None
        else:
            newnode.next = self.head
            self.head = newnode

    def InsertAtEnd(self,data):
        newnode=node(data)
        if self.head == None:
            self.head=newnode
            newnode.next=None
        else:
            temp=self.head
            while(temp.next != None):
                temp=temp.next
            newnode.next=None
            temp.next=newnode

    def display(self):
        if self.head == None:
            print("Empty!!")
        else:
            temp = self.head
            print("Linked list: ", end=" ")
            while temp != None:
                print(temp.data, end="->")
                temp = temp.next
            print()

ll = linkedlist()
choice = 1
while choice != 0:
    print("1.Add node at beg")
    print("2.Add node at end")
    print("3.Display")
    print("0.Exit")
    choice = int(input("Enter your choice:"))
    if choice == 1:
        data = int(input("Enter data of a choice:"))
        ll.InsertAtbeg(data)
    elif choice == 2:
        data=int(input("Enter data of a choice:"))
        ll.InsertAtEnd(data)
    elif choice == 3:
        ll.display()
    elif choice == 0:
        print("Bye!")
    else:
       print("Invalid choice")
```

# Stack

```python
class stack():
    size=5
    top=-1
    stack=[0]*size

    def overflow(self):
        if(self.top == self.size-1):
            return True
        return False
    
    def underflow(self):
        if(self.top == -1):
            return True
        return False

    def push(self,data):
        if(self.overflow()):
            print("Overflow")
        else:
            self.top+=1
            self.stack[self.top]=data

    def pop(self):
        if(self.underflow()):
            print("Underflow")
        else:
            self.top-=1

    def display(self):
        for i in range(self.top,-1,-1):
            print(self.stack[i])

        
s=stack()
choice=1
while choice!=0:
    print("1.Push()")
    print("2.Pop()")
    print('3.Display')
    choice=int(input("Enter a choice:"))
    if choice==1:
        data=int(input("Enter data:"))
        s.push(data)
    elif choice==2:
        s.pop()
    elif choice==3:
        s.display()
    elif choice==0:
        print("Bye!!")
    else:
        print("Invalid choice")
```
