# Sorting

## Insertion Sorting

```python
n=int(input())
l=list(map(int,input().split()))

print("Original list:",l)
for i in range(1,n):
    temp=l[i]
    j=i-1
    while j>=0 and temp<=l[j]:
        l[j+1]=l[j]
        j-=1
    l[j+1]=temp

print("Updated list:",l)
```

### Program to count vowel and consonant

```python
s=input()
print("String:",s)
vowel="aieouAEIOU"
v_c=0
c_c=0
for i in s:
    if i in vowel:
        v_c+=1
    else:
        c_c+=1

print("Vowel count:",v_c)
print("Consonant count:",c_c)
```

## Quick Sort

```python
def quick(arr,low,high):
    if low < high:
        part=partition(arr,low,high)
        quick(arr,low,part-1) # left subarray
        quick(arr,part+1,high) # right subarray
    return arr

def partition(arr,low,high):
    i=low
    j=i-1
    pivot=arr[high]
    while(i<=high-1):
        if(arr[i]<=pivot):
            j+=1
            # swap arr[i]=arr[j]
            temp=arr[i]
            arr[i]=arr[j]
            arr[j]=temp
            i+=1
        else:
            i+=1
    # swap arr[j+1]=arr[high]
    temp=arr[j+1]
    arr[j+1]=arr[high]
    arr[high]=temp
    return j+1    

n=int(input())
arr=list(map(int,input().split()))
print("Original list:",arr)
print("Updated list:",quick(arr,0,n-1))
```

## Linkedlist

### Single linked-list

```python
class node:
    def __init__(self, data):
        self.data = data
        self.next = None

class linkedlist:
    def __init__(self):
        self.head = None

    def InsertAtbeg(self, data):
        newnode = node(data)
        if self.head == None:
            self.head = newnode
        else:
            newnode.next = self.head
            self.head = newnode

    def display(self):
        if self.head == None:
            print("Empty!!")
        else:
            temp = self.head
            print("Linked list: ", end=" ")
            while temp != None:
                print(temp.data, end="->")
                temp = temp.next
            print()

ll = linkedlist()
choice = 1
while choice != 0:
    print("1.Add node at beg")
    print("2.Display")
    print("0.Exit")
    choice = int(input("Enter your choice:"))
    if choice == 1:
        data = int(input("Enter data of a choice:"))
        ll.InsertAtbeg(data)
    elif choice == 2:
        ll.display()
    # elif choice == 0:
    # #     print("Bye!")
    # # else:
    # #    print("Invalid choice")
```
