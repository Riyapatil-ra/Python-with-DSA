# Binary Tree

* Inorder traversal: **Left-Root-Right**
* Preorder traversal: **Root-Left-Right**
* Postorder traversal: **Left-Right-Root**

```python
from collections import deque
class node:
    def __init__(self,data):
        self.left=None
        self.data=data
        self.right=None

class Tree:
    def __init__(self):
        self.root=None

    def insert(self,data):
        newnode=node(data)
        
        if self.root==None:
            self.root=newnode
            return
        q=deque([self.root])
        while q:
            temp=q.popleft()

            if temp.left==None:
                temp.left=newnode
                return
            else:
                q.append(temp.left)
        
            if temp.right==None:
                temp.right=newnode
                return
            else:
                q.append(temp.right)

    def bfs(self):
        if self.root==None:
            print("empty")
            return
        q=deque([self.root])
        while q:
            temp=q.popleft()
            print(temp.data,end=",")
            if temp.left!=None:
                q.append(temp.left)
            if temp.right!=None:
                q.append(temp.right)
        
    def dfs(self):
        if self.root==None:
            print("empty")
            return
        l=[self.root]
        while l:
            temp=l.pop()
            print(temp.data,end=",")
            if temp.right!=None:
                l.append(temp.right)
            if temp.left!=None:
                l.append(temp.left)

    def inorder(self,root): # left-root-right
        if(root==None):
            return
        self.inorder(root.left)
        print(root.data,end=',')
        self.inorder(root.right)
    
    def preorder(self,root): # root-left-right
        if(root==None):
            return
        print(root.data,end=',')
        self.preorder(root.left)
        self.preorder(root.right)
    
    def postorder(self,root): # left-right-root
        if(root==None):
            return
        self.postorder(root.left)
        self.postorder(root.right)    
        print(root.data,end=',')           
        
t=Tree()
t.insert(10)
t.insert(20)
t.insert(50)
t.insert(30)
t.insert(40)
t.insert(60)

print("Inorder Traversal:")
t.inorder(t.root)
print("\n\nPreorder traversal:")
t.preorder(t.root)
print("\n\nPostorder traversal:")
t.postorder(t.root)
print("\n\nBFS Traversal:")
t.bfs()
print("\n\nDFS Traversal:")
t.dfs()
```

# Circular Queue

* Time complexity of Enqueue operation is **O(1)**
* Time complexity of Dequeue operation is **O(1)**
* The Main advantage of circular queue over linear queue is **Better memory Utilization**
* **Dequeue** operation removes elements from circular queue
* **Enqueue** operation add elements in circular queue
* A circular queue is empty when **front==-1**
* A circular queue is full when **rear+1%size==front**

## Factorial program

```python
def fact(num):
    if num == 0:
        return 1
    elif num == 1:
        return num
    else:
        return num*fact(num-1)

num=int(input("Enter number:"))
print(f"Factorial of {num} = {fact(num)}")
```

## Anagram program

```python
s=input("Enter words:").split() # create a list of words using split() method. abc bac cab car = [abc, bac, cab, car]
d={}                             # create a dictionary to store the anagrams
for i in s:
    key="".join(sorted(i)) # 1st sort the word alphabetically in list and then join it to form string.[abc, bac, rac, car] = [abc, abc, acr, acr]
    if key not in d:
        d[key]=[]
    d[key].append(i)

for i in d.values():
    print(*i) # using * operator to unpack the list and print the words in a single line.[abc, bac] [rac, car] = abc bac rac car
```
